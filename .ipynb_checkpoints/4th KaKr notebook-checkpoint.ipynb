## Feature 설명

- This data was extracted from the 1994 Census bureau database by Ronny Kohavi and Barry Becker (Data Mining and Visualization, Silicon Graphics). A set of reasonably clean records was extracted using the following conditions: ((AAGE>16) && (AGI>100) && (AFNLWGT>1) && (HRSWK>0)). The prediction task is to determine whether a person makes over "$50K" a year.
    
    - age : 나이
    - workclass : 고용 형태
    - fnlwgt : 사람 대표성을 나타내는 가중치 (final weight의 약자)
            - A single cell estimate of the population 16+ for each state.
            - Controls for Hispanic Origin by age and sex.
            - Controls by Race, age and sex.
            -CPS sample is actually a collection of 51 state samples, each with its own probability of selection, the statement only applies within state.
    - education : 교육 수준
    - education_num : 교육 수준 수치
    - marital_status: 결혼 상태
    - occupation : 업종
    - relationship : 가족 관계
    - race : 인종
    - sex : 성별
    - capital_gain : 양도 소득
    - capital_loss : 양도 손실
    - hours_per_week : 주당 근무 시간
    - native_country : 국적
    - income : 수익 (예측해야 하는 값)
             >50K : 1
             <=50K : 0

```python
# def get_keys(df):
#     return list(df.keys())

def summarizing_df(df):
    
    print("\n1. Keys")
    print(df.columns)
    print("\n2. Shape")
    print(df.shape)
    print("\n3. Head")
    print(df.head(5))

def lable_encode_df(df):
    le = LabelEncoder()
    for col in df.columns:
        print(f'Column [{col}] is "{df[col].dtypes}"')
        if df[col].dtypes == 'object':
            print(f'\tencoding [{col}]....')
            df[col] = le.fit_transform(df[col])
    print('[Label encoding done.]')
    print("\n1. Keys")
    print(df.columns)
    print("\n2. Shape")
    print(df.shape)
    print("\n3. Head")
    print(df.head(5))

    return df

def plot_histogram(df):

    plt.figure(figsize=(15, 10))
    for idx_k, key in enumerate(df.columns):
        plt.subplot(4, 5, idx_k+1)
        plt.title(key)
        df[key].hist()
        plt.grid()
    plt.tight_layout()
    plt.show()
```

```python
# id Drop
train.drop(['id'], axis=1, inplace=True)
test.drop(['id'], axis=1, inplace=True)
```

### LGBM Clf

```python
lgbm_clf = LGBMClassifier(n_estimators=4000,
                          learning_rate=0.03,
                          min_child_samples=20,
                          num_leaves= 30,
                          max_depth= 50,
                          random_state=1105)
start_time = time.time()

lgbm_clf.fit(X_train,
             y_train,
             early_stopping_rounds=100,
             eval_metric='f1',
             eval_set=[(X_train, y_train),(X_test, y_test)],
             verbose=True
            )
lgbm_pred = lgbm_clf.predict(X_test)
lgbm_f1 = f1_score(y_test, lgbm_pred)
lgbm_accuracy = accuracy_score(y_test, lgbm_pred)

print('LGBM 정확도: {0:.4f}'.format(lgbm_accuracy))
print(f'LGBM f1 score: {lgbm_f1:.4f}')
print('LGBM 수행 시간: {0:.1f} 초'.format(time.time() - start_time))
```

```python
# LGBM Regressor params
params_full = {
    'n_estimators': [2000, 4000],
    "num_leaves": [10, 15, 20, 30],
    "learning_rate": [0.03, 0.05, 0.1],
    "bagging_fraction": [0.5, 0.6, 0.7, 0.8],
    "bagging_seed": [0],
    "num_threads": [2, 4, 8],
    "colsample_bytree": [0.7],
    # "device" : "gpu"
}

params = {
    'n_estimators': [2000, 4000],
    'learning_rate': [0.03],
    # 'min_child_samples' : [1, 3, 5, 20],
    'num_leaves': [20, 30],
    #     "num_threads": [4],
    'max_depth': [10, 15, 20],
    "bagging_fraction": [0.5, 0.6],
}

params2 = {'n_estimators': [4000],
           'learning_rate': [0.03],
           'min_child_samples': [20],
           'num_leaves': [12,30],
           'max_depth': [50],
           'random_state': [1105]}


fit_params = {
#     "early_stopping_rounds": [100],
    "eval_metric": "acc",
    "eval_set": [[X_test, y_test]]
}
```

```python
# LGBM Regressor + GridSearchCV
start_time = time.time()

grid_cv = GridSearchCV(lgbm_clf, param_grid = params2, cv = 5 ,verbose = False, n_jobs = -1)
grid_cv.fit(X_train, y_train, **fit_params)
```

```python
# grid_cv
```

```python
print('최적 하이퍼 파라미터:\n', grid_cv.best_params_)
print('최고 예측 정확도: {0:.4f}'.format(grid_cv.best_score_))
print("LGBM 수행 시간: {0:.1f} 초 ".format(time.time() - start_time))

model = grid_cv.best_estimator_
predictions = model.predict(X_test)
grid_f1 = f1_score(y_test, predictions)
acc = accuracy_score(y_test, predictions)

print('테스트 세트에서의 LGBM acc: {0:.4f}'.format(acc))
print('테스트 세트에서의 LGBM f1 score: {0:.4f}'.format(grid_f1))
```
